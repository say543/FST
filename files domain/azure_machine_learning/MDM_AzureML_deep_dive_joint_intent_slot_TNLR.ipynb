```python
##BATCH AI
from azureml.train.dnn import PyTorch



### for debug if data is reable
from azureml.core import Dataset



training_datastore_path = [(ds, './azureml_data/MDM_TrainSet_01202021v1.tsv')]
#training_datastore_path = [(ds, './azureml_data/MDM_TrainSet_small_01202021v1.tsv')]


training_dataset_obj = Dataset.File.from_files(path=training_datastore_path)

training_dataset_name = 'training_non_contexual_slot_data_set'
training_dataset_description = 'training_non_contexual_slot_data_set description'


training_dataset = training_dataset_obj.register(workspace=ws, name=training_dataset_name, description=training_dataset_description, create_new_version=True)


# augmented data test 
augmented_datastore_path = [(ds, './azureml_data/open_ppt_augmentation.tsv')]

augmented_dataset_obj = Dataset.File.from_files(path=augmented_datastore_path)

augmented_dataset_name = 'augmented_non_contexual_slot_data_set'
augmented_dataset_description = 'augmented_non_contexual_slot_data_set description'


augmented_dataset = augmented_dataset_obj.register(workspace=ws, name=augmented_dataset_name, description=augmented_dataset_description, create_new_version=True)


# tokenizer enforcement set
tokenizer_enforcement_datastore_path = [(ds, './azureml_data/tokenizer_enforcement_query.tsv')]

tokenizer_enforcement_dataset_obj = Dataset.File.from_files(path=tokenizer_enforcement_datastore_path)

tokenizer_enforcement_dataset_name = 'tokenizer_enforcement_non_contexual_slot_data_set'
tokenizer_enforcement_dataset_description = 'tokenizer_enforcement_non_contexual_slot_data_set description'


augmented_dataset = tokenizer_enforcement_dataset_obj.register(workspace=ws, name=tokenizer_enforcement_dataset_name, description=tokenizer_enforcement_dataset_description, create_new_version=True)


#TNLR model
TNLRv3_path = [(ds, './TNLR/tnlrv3-base.pt')]

TNLRv3_obj = Dataset.File.from_files(path=TNLRv3_path)

TNLRv3_name = 'TNLRv3_pretrained_model'
TNLRv3_description = 'TNLRv3_pretrained_model description'

TNLRv3_model = TNLRv3_obj.register(workspace=ws, name=TNLRv3_name, description=TNLRv3_description, create_new_version=True)



# for debug new format
import pandas as pd
file_name = training_dataset.download()[0]
df = pd.read_csv(file_name, sep='\t', encoding="utf-8")
print(df.head())





from azureml.core.runconfig import MpiConfiguration
from azureml.train.hyperdrive import GridParameterSampling
from azureml.train.hyperdrive import HyperDriveConfig
from azureml.train.hyperdrive import PrimaryMetricGoal
from azureml.train.hyperdrive.parameter_expressions import choice


script_folder = './training_script'



# new dataset name
script_params = {
    #'--dataset_name': 'training_data_set'
    '--dataset_name': 'training_non_contexual_slot_data_set',
    #augment dataset name
    #'--augment_dataset_name': 'augmented_non_contexual_slot_data_set',
    #  tokenizer enforcement dataset
    '--tokenizer_enforcement_dataset_name' : 'tokenizer_enforcement_non_contexual_slot_data_set',
    '--TNLR_model_bin': 'TNLRv3_model',
    #'--pretrain_model': 'distilbert-base-uncased'
    '--pretrain_model': 'bert-base-uncased'
    #'--pretrain_model': 'distilbert-base-uncased'
}



estimator = PyTorch(source_directory = script_folder,
                    compute_target = compute_target,
                    entry_script = 'MDM_train_horovod_joint_intent_slot_TNLR.py',
                    script_params = script_params,
                    use_gpu = True,
                    #node_count=8,
                    # for 
                    #node_count=6,
                    #node_count=4,
                    node_count=2,
                    #node_count=1,
                    process_count_per_node=1,
                    distributed_training=MpiConfiguration(),
                    # setup transformers package number
                    #pip_packages = ['sklearn', 'joblib', 'transformers', 'azureml-dataprep[fuse,pandas]'])
                    pip_packages = ['sklearn', 'joblib', 'transformers==4.1.1', 'azureml-dataprep[fuse,pandas]', 'seqeval'])


#ve
param_sampling = GridParameterSampling({
    'batch_size': choice(32),
    # originla learning rate
    #'learning_rate': choice(1e-5),
    # change to bigger learning rate
    'learning_rate': choice(5e-5),
    # combined tow at the same time
    #'learning_rate': choice(5e-5, 1e-5),
    'adam_epsilon': choice(1e-8),
    'seed_val': choice(111),
    # add epoch parameter
    #'num_epochs': choice(1)})
    #'num_epochs': choice(2)})
    #'num_epochs': choice(5)})
    #'num_epochs': choice(25)})
    'num_epochs': choice(10)})
    #'num_epochs': choice(10)})
    #'num_epochs': choice(5,10)})
    #'num_epochs': choice(5,10,15)})
    #'num_epochs': choice(15,20,25)})
    #'num_epochs': choice(5,10,15,20,25)})



#horovod
#https://docs.microsoft.com/en-us/azure/machine-learning/how-to-train-pytorch
hyperdrive_run_config = HyperDriveConfig(estimator = estimator,
                                         hyperparameter_sampling = param_sampling,
                                         policy = None,
                                         primary_metric_name = 'validation loss',
                                         primary_metric_goal = PrimaryMetricGoal.MINIMIZE,
                                         # increase to 3 to see different validation through graph
                                         # ? but graph still only outputs one point
                                         #max_total_runs = 1,
                                         max_total_runs = 5,
                                         max_concurrent_runs = 1)


#hyperdrive_run = exp.submit(hyperdrive_run_config)
hyperdrive_run = experiment.submit(hyperdrive_run_config)
# show detail
from azureml.widgets import RunDetails
RunDetails(hyperdrive_run).show()


# this is to show metric so comment it at first

hyperdrive_run.wait_for_completion(show_output = False)

children = list(hyperdrive_run.get_children())
metricslist = {}
i = 0

for single_run in children:
    results = {k: v for k, v in single_run.get_metrics().items() if isinstance(v, float)}
    parameters = single_run.get_details()['runDefinition']['arguments']
    try:
        results['batch_size'] = parameters[3]
        results['learning_rate'] = parameters[5]
        results['adam_epsilon'] = parameters[7]
        results['num_epochs'] = parameters[9]
        metricslist[i] = results
        i += 1
    except:
        pass

rundata = pd.DataFrame(metricslist).sort_index(1).T.sort_values(by = ['validation loss'], ascending = True)
rundata


```
